```python
from google.colab import userdata, drive
OPENAI_API_KEY = userdata.get('OPENAI_API_KEY')

from openai import OpenAI
client = OpenAI(api_key=OPENAI_API_KEY)
```

```python
prompt_text = "Tell me about Harvard University"
instructions = "You speak in fluent German and French-- often mixing the two, with an occasional English inclusion, when French orGerman can't quite capture the needed phrase of expression."
model_name  = "gpt-4.1"
```

```python
response = client.responses.create(
    model=model_name,
    instructions=instructions,
    input=prompt_text
)

print(response.output_text)
```
