```python
from google.colab import userdata, drive
OPENAI_API_KEY = userdata.get('OPENAI_API_KEY')
drive.mount("/content/drive")

from openai import OpenAI
client = OpenAI(api_key=OPENAI_API_KEY)
```

```python
image_prompt = "Create a visual interpretation of Jorge Luis Borgesâ€™ poem 'Otro poema de los dones', designed for an English-speaking audience, using symbolic imagery that reflects its themes of gratitude, libraries, and the infinite."
```

```python
response = client.images.generate(
    model="dall-e-3",
    prompt=image_prompt,
    n=1,
    size="1024x1024",
    response_format="b64_json"
)

image_b64 = response.data[0].b64_json
img_bytes = base64.b64decode(image_b64)
img = Image.open(io.BytesIO(img_bytes))
display(img)
```

## saving the image to an output folder in gdrive

```python

img_dir = Path(
    "/content/drive/Shareddrives/LL_CENTRAL/LL_PROJECTS/24.1/COLAB/ll_color_colabs/_ai_lab/utilities/output-folder/images"
)

img_dir.mkdir(parents=True, exist_ok=True)

stem_prompt = ''.join(w.capitalize() for w in re.findall(r'\w+', image_prompt))[:60]
timestamp   = datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
filename    = f"{stem_prompt}_{timestamp}.png"

(img_dir / filename).write_bytes(img_bytes)

print(f"Image saved to:\n    {img_dir / filename}")
```
