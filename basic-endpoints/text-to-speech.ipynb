```python
from google.colab import userdata, drive
OPENAI_API_KEY = userdata.get("OPENAI_API_KEY")
drive.mount("/content/drive")

from openai import OpenAI
client = OpenAI(api_key=OPENAI_API_KEY)
```

```python
tts_text   = "The quick brown fox jumped over the lazy dog."
model_name = "gpt-4o-mini-tts"
voice_name = "alloy"
out_path   = "/content/drive/Shareddrives/LL_CENTRAL/LL_PROJECTS/24.1/COLAB/ll_color_colabs/_ai_lab/utilities/output-folder/audio"
```

```python
speech_file = Path(out_path)

with client.audio.speech.with_streaming_response.create(
        model=model_name,
        voice=voice_name,
        input=tts_text
    ) as response:
    response.stream_to_file(speech_file)

print(f"Audio saved to: {speech_file}")
display(Audio(str(speech_file)))
```
