## Define functions to carry out "ping pong"


*   First function: text completion that carries out a "free" translation; no specified elements of the text to take into account.
*   Second and third functios: recto and verso; carry out translation back and forth between the languages.



```python
client = OpenAI(api_key=openai.api_key)

def translatebasic(outputlang, inputtext, systemprompt):
    response = client.responses.create(
        model="gpt-4o",
        input=f"{systemprompt}\nTranslate the following text to {outputlang}. Text: {inputtext}.",
    )
    final = response.output_text
    return(final)

def recto(text):
    english = translatebasic("english", text, "") # Added a placeholder for systemprompt
    return english

def verso(text):
    chinese = translatebasic("chinese", text, "") # Added a placeholder for systemprompt
    return chinese
```

## Carry out "ping pong":

*  Define number of trials for ping pong (one trial is one back and one forth).
*   Carry out use a for loop to test for convergence.

```python
trials = 100
```

```python
text0 = confucius
```

```python
for i in range(trials):
  text1 = recto(text0)
  print(text1)
  text0 = verso(text1)
  print(text0)
```
