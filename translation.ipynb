```python
from google.colab import userdata, drive
OPENAI_API_KEY = userdata.get("OPENAI_API_KEY")
drive.mount("/content/drive")

from openai import OpenAI
client = OpenAI(api_key=OPENAI_API_KEY)
```

```python
input_mode      = "text"                 # "audio"  or  "text"

input_language = "German"
output_language = "English"


audio_path      = "/content/drive/Shareddrives/LL_CENTRAL/LL_PROJECTS/24.1/COLAB/ll_color_colabs/_ai_lab/utilities/input-folder/audio/speech.mp3"
audio_model     = "whisper-1"

# OR:

input_text     = "Guten Tag! Wie geht es Ihnen heute?"
text_model      = "gpt-4.1"
```

```python
def gpt_translate(text, tgt_lang):
    prompt = f"Translate the following text into {output_language}:\n\n{text}"
    resp   = client.responses.create(
        model=text_model,
        input=prompt,
    )
    return resp.output_text.strip()

if not input_language or input_language == "[language]" or not output_language or output_language == "[language]":
    raise ValueError("Please set input_language and output_language to the desired languages.")


if input_mode == "audio":
    with open(audio_path, "rb") as f:
        if output_language.lower() in {"en", "english"}:
            result = client.audio.translations.create(
                model=audio_model,
                file=f,
                response_format="json",
                prompt=None,
            )
            translated_text = result.text.strip()
        else:
            transcript = client.audio.transcriptions.create(
                model=audio_model,
                file=f,
                language=input_language,
                response_format="json"
            ).text.strip()
            translated_text = gpt_translate(transcript, output_language)

    print(f"ðŸ”Š  Audio translated to {output_language}:\n")
    print(translated_text)

elif input_mode == "text":
    translated_text = gpt_translate(input_text, output_language)
    print(f"Text translated to {output_language}:\n")
    print(translated_text)

else:
    raise ValueError("input_mode must be 'audio' or 'text'")
```
